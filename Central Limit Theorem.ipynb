```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
n = 10
p = 0.1
size = 1000

binomial_dist = np.random.binomial(n, p, size)
plt.hist(binomial_dist, density=True)
plt.show()
```

```python
num_samples = 10000
sample_size = 300

distribution_range = (0, 1)

samples = np.random.uniform(distribution_range[0],distribution_range[1],(num_samples, sample_size))

# calculate the sample mean
sample_mean = np.mean(samples, axis=1)

plt.hist(sample_mean, bins=30, density = True, edgecolor='black')
plt.title('Histogram of sample means')
plt.xlabel('Sample Mean')
plt.ylabel("Density")
plt.show()
```

```python
no_sample = 1000
sample_size=50
lambda_param = 2

samples = np.random.exponential(scale=1/lambda_param, size=(no_sample, sample_size))
# calculate the sample mean
sample_mean = np.mean(samples, axis=1)

plt.hist(sample_mean, bins=30, density = True, edgecolor='black')
plt.title('Histogram of sample means')
plt.xlabel('Sample Mean')
plt.ylabel("Density")
plt.show()
```
