```python
# Extract year
dataset['year'] = dataset['date'].dt.year
```

```python
dataset.head(10)
```

```python
# Extract Month
dataset['Month'] = dataset['date'].dt.month
dataset.head()
```

```python
# Extract Days
dataset['Days'] = dataset['date'].dt.day
dataset.head()
```

```python
# Month Name
dataset['Month_Name'] = dataset['date'].dt.month_name()
dataset.head()
```

```python
# Days name
dataset['Days_Name'] = dataset['date'].dt.day_name()
dataset.head()
```

```python
# is weekend or weekday ?

dataset['date_is_weekend'] = np.where(dataset['Days_Name'].isin(['Sunday','Saturday']),'weekend','weekday')
```

# Titanic Dataset - Handling missing value

```python
titanic = pd.read_csv('titanic_train.csv')
titanic.head()
```

```python
titanic.info()
```

```python
# Check missing data
titanic.isnull().sum()
```

```python
# Check missing percentage data
titanic.isnull().sum() / len(titanic)*100
```

```python
# remove Cabin variable as there are 77.1% missing data found
titanic = titanic.drop(['Cabin'], axis=1)
```

```python
titanic.isnull().sum() / len(titanic)*100
```

```python
# Embarked - Object : Impute (Mode)
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Embarked'].mode()
```

```python
titanic['Embarked'] = titanic['Embarked'].fillna('S')
# titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode())
```

```python
titanic.isnull().sum() / len(titanic)*100
```

```python
# Age - Float value (first we have to check outlier and then will decide whether we have to use Mean or Median)
```

```python
# Check outlier
import seaborn as sns
sns.boxplot(y='Age', data=titanic)
```

```python
titanic['Age'].median()
```

```python
titanic['Age'].mean()
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
```

```python
titanic.isnull().sum() / len(titanic)*100
```
